2. Given the following scenarios, classify each as supervised, unsupervised, or reinforcement learning: 1) Netflix recommending movies based on your watch history, 2) Spotify grouping similar songs into playlists, 3) A self-driving car learning to park by trial and error. Write your answers with a one-line explanation for each choice.



1.Netflix recommending movies based on your watch history

Answer: Supervised Learning

Explanation: The system learns from users' past viewing behavior to predict which movies they are likely to enjoy.

2.Spotify grouping similar songs into playlists

Answer: Unsupervised Learning

Explanation: It groups songs with similar characteristics without using labeled categories.

3.A self-driving car learning to park by trial and error

Answer: Reinforcement Learning

Explanation: The car learns the best parking strategy by receiving rewards for successful actions and penalties for mistakes.

4. Imagine you are building a spam filter for WhatsApp messages. Write a short paragraph explaining how traditional programming would approach this problem versus how machine learning would approach it.<br><br><em><strong>Hint:</strong> Focus on rules vs learning from data.</em>

Traditional programming would detect spam by using predefined rules. For example, if a message contains words like "Win Money" or includes too many links, it is marked as spam. However, these rules must be manually created and updated whenever new spam patterns appear.

Machine learning, on the other hand, learns from a large dataset of labeled messages (spam and non-spam). Instead of relying on fixed rules, it identifies patterns automatically and predicts whether a new message is spam. As more data becomes available, the model can improve its accuracy and adapt to new types of spam without requiring manual rule updates.
