1. Given two 2x2 matrices representing the number of likes and comments on two Instagram posts over two days, write Python code to multiply them and display the resulting matrix.


```python
import numpy as np

A = np.array([
    [120, 150],
    [180, 200]
])

B = np.array([
    [10, 15],
    [20, 25]
])

result = np.dot(A, B)

print("Likes Matrix (A):")
print(A)

print("\nComments Matrix (B):")
print(B)

print("\nResult of Matrix Multiplication (A × B):")
print(result)
```

2. Create a simple Python function linear_regression_predict(x, m, c) that takes an input value x, slope m, and intercept c, and returns the predicted output y for a linear regression equation y = mx + c.


y=mx+c

where:

1. x = Input value
2. m = Slope
3. c = Intercept
4. y = Predicted output

```python
def linear_regression_predict(x, m, c):
    y = m * x + c
    return y

x = 10
m = 2.5
c = 5

prediction = linear_regression_predict(x, m, c)

print("Input (x):", x)
print("Slope (m):", m)
print("Intercept (c):", c)
print("Predicted Output (y):", prediction)
```

3. Imagine you are predicting the delivery time for Zomato orders based on distance. Given three distances (2km, 5km, 8km) and a linear regression equation y = 4x + 10 (where y is minutes), use your function from the previous task to calculate the predicted delivery time for each distance.


y=4x+10

where:

x = Distance (km)
y = Predicted delivery time (minutes)

```python
def linear_regression_predict(x, m, c):
    return m * x + c

m = 4
c = 10

distances = [2, 5, 8]

for distance in distances:
    predicted_time = linear_regression_predict(distance, m, c)
    print(f"Distance: {distance} km -> Predicted Delivery Time: {predicted_time} minutes")
```

4. Explain in your own words what the slope and intercept represent in the context of Flipkart's product price prediction (e.g., predicting price based on product rating). Write 2-3 sentences for each term.


Slope (m)

The slope represents how much the predicted product price changes when the product rating increases by 1 point. If the slope is positive, higher-rated products are expected to have higher prices. If it is negative, higher ratings are associated with lower predicted prices.

Intercept (c)

The intercept is the predicted product price when the product rating is 0. Although a rating of 0 may not be realistic on Flipkart, the intercept serves as the starting value of the linear regression model and helps define the prediction equation.
