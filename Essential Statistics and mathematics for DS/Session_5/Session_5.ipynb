2. Write a Python script that simulates random sampling: given a list of 1000 fake user IDs (e.g., user1 to user1000), select 50 random users to simulate a survey sample. Print the selected user IDs.<br><br><em><strong>Hint:</strong> Use the random.sample() function from Python's random module.</em>


```python
import random

users = [f"user{i}" for i in range(1, 1001)]

sample_users = random.sample(users, 50)

print("Selected Survey Sample (50 Users):")
for user in sample_users:
    print(user)
```

4. Given this scenario: A survey about music preferences is only sent to users who recently streamed Bollywood songs on Spotify. Identify and explain the type of bias this introduces in the sample.


**Type of Bias:** **Selection Bias (Sampling Bias)**

**Explanation:**
The survey is sent only to users who recently streamed **Bollywood songs**, so the sample is not representative of all Spotify users. People who mainly listen to pop, rock, hip-hop, classical, or other genres are excluded. As a result, the survey results may be biased toward Bollywood listeners and may not accurately reflect the music preferences of the entire Spotify user base.


5. Choose any two types of sampling (for example: random sampling, stratified sampling, systematic sampling, cluster sampling). For each, describe how you would use it to select a sample of users from BookMyShow to study movie booking habits.

### 1. Random Sampling

In **random sampling**, every BookMyShow user has an equal chance of being selected. For example, randomly select **500 users** from all registered users to study their movie booking habits. This helps reduce bias and provides a representative sample.

### 2. Stratified Sampling

In **stratified sampling**, users are divided into groups (strata), such as **age groups** (18–25, 26–40, 41+). Then, randomly select users from each age group in proportion to their size. This ensures that all age groups are represented in the study.
