2. Given the following array of delivery times (in minutes) for Zomato orders: [30, 25, 40, 35, 30, 45, 30], write code to find the median delivery time and print it.


```python
delivery_times = [30, 25, 40, 35, 30, 45, 30]

delivery_times.sort()

n = len(delivery_times)

if n % 2 == 1:
    median = delivery_times[n // 2]
else:
    mid1 = delivery_times[n // 2 - 1]
    mid2 = delivery_times[n // 2]
    median = (mid1 + mid2) / 2

print("Sorted Delivery Times:", delivery_times)
print("Median Delivery Time:", median)
```

3. Write a function most_common_rating(ratings) that takes a list of Flipkart product ratings (e.g., [5, 4, 4, 3, 5, 4, 2, 4]) and returns the mode (most frequent rating).


```python
def most_common_rating(ratings):
    frequency = {}

    # Count frequency of each rating
    for rating in ratings:
        if rating in frequency:
            frequency[rating] += 1
        else:
            frequency[rating] = 1

    mode = max(frequency, key=frequency.get)

    return mode


ratings = [5, 4, 4, 3, 5, 4, 2, 4]

result = most_common_rating(ratings)

print("Product Ratings:", ratings)
print("Most Common Rating (Mode):", result)
```

4. Given three lists representing YouTube video views for three different channels, compare the mean, median, and mode for each channel, and decide which channel's data is most affected by outliers.<br><br><em><strong>Hint:</strong> Try using one list with a very high value (e.g., [100, 120, 110, 105, 5000]) to see the effect on mean vs median.</em>


```python
import statistics

channel_A = [100, 120, 110, 105, 5000]   
channel_B = [200, 220, 210, 230, 240]
channel_C = [50, 60, 55, 65, 70]

def analyze_channel(data):
    mean = statistics.mean(data)
    median = statistics.median(data)

    try:
        mode = statistics.mode(data)
    except:
        mode = "No unique mode"

    return mean, median, mode


for name, data in [("Channel A", channel_A),
                   ("Channel B", channel_B),
                   ("Channel C", channel_C)]:

    mean, median, mode = analyze_channel(data)

    print(name)
    print("Views:", data)
    print("Mean:", mean)
    print("Median:", median)
    print("Mode:", mode)
```

5. Pick any real dataset from Kaggle or UCI (e.g., IPL player runs, Swiggy restaurant ratings, or movie box office collections), load it using pandas, and display the mean, median, and mode for one numeric column of your choice.

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("ipl_players.csv")

print(df.head())

runs = df["Runs"]

mean_runs = runs.mean()
median_runs = runs.median()
mode_runs = stats.mode(runs, keepdims=True).mode[0]

print("Mean Runs:", round(mean_runs, 2))
print("Median Runs:", median_runs)
print("Mode Runs:", mode_runs)
```
