1. Write a Python function bayes_posterior(prior, likelihood, evidence) that calculates and returns the posterior probability using Bayes theorem formula for given values.


P(A∣B)= P(B∣A)×P(A)/p(B)


```python
def bayes_posterior(prior, likelihood, evidence):
    posterior = (likelihood * prior) / evidence
    return posterior

prior = 0.3
likelihood = 0.8
evidence = 0.5

result = bayes_posterior(prior, likelihood, evidence)

print("Posterior Probability:", result)
```

2. Given the following scenario: On Swiggy, 10% of restaurants are 'premium', and 80% of premium restaurants offer a discount, while only 20% of regular restaurants do. If you see a discount, use Bayes theorem in code to find the probability that the restaurant is premium.<br><br><em><strong>Hint:</strong> Define prior, likelihood, and evidence in your code with comments.</em>


```python
def bayes_posterior(prior, likelihood, evidence):
    return (prior * likelihood) / evidence

prior = 0.10

likelihood = 0.80

regular = 0.90

regular_discount = 0.20

evidence = (likelihood * prior) + (regular_discount * regular)

posterior = bayes_posterior(prior, likelihood, evidence)

print("Evidence:", evidence)
print("Probability that the restaurant is Premium given a discount:", posterior)
```

3. Simulate a basic spam detection filter: Given a list of 20 email subjects (some spam, some not), and the probability that 'free' appears in spam vs non-spam emails, write code to calculate the probability an email is spam if it contains 'free', using Bayes theorem.


```python
emails = [
    "Free Gift Inside",
    "Meeting Reminder",
    "Win a Free iPhone",
    "Project Update",
    "Free Discount Coupon",
    "Invoice Attached",
    "Congratulations You Won",
    "Free Vacation Offer",
    "Team Meeting Today",
    "Limited Time Free Trial",
    "Salary Credited",
    "Urgent Account Update",
    "Free Cashback Offer",
    "Your Order Shipped",
    "Free Recharge Now",
    "Weekly Newsletter",
    "Claim Your Prize",
    "Free Membership",
    "Interview Schedule",
    "Exclusive Free Bonus"
]

def bayes_posterior(prior, likelihood, evidence):
    return (prior * likelihood) / evidence

prior_spam = 0.40

likelihood = 0.80

free_given_nonspam = 0.10

prior_nonspam = 1 - prior_spam

evidence = (likelihood * prior_spam) + (free_given_nonspam * prior_nonspam)

posterior = bayes_posterior(prior_spam, likelihood, evidence)

print("Email Subjects:")
for email in emails:
    print("-", email)

print("\nProbability an email is Spam if it contains 'free':")
print(round(posterior, 4))
print("Percentage:", round(posterior * 100, 2), "%")
```

"Generate a real-world Bayes Theorem example related to Flipkart. Assume 15% of products are on a special sale, 90% of sale products display a 'Sale' badge, and 20% of non-sale products also display the badge. Write Python code to calculate the probability that a product is actually on sale if it shows the 'Sale' badge."

```python
def bayes_posterior(prior, likelihood, evidence):
    return (prior * likelihood) / evidence

prior_sale = 0.15

likelihood = 0.90

badge_given_nonsale = 0.20

prior_nonsale = 1 - prior_sale

evidence = (likelihood * prior_sale) + (badge_given_nonsale * prior_nonsale)

posterior = bayes_posterior(prior_sale, likelihood, evidence)

print("Prior Probability (Sale):", prior_sale)
print("Likelihood:", likelihood)
print("Evidence:", evidence)
print("Probability product is on sale given it shows 'Sale' badge:")
print(round(posterior, 4))
print("Percentage:", round(posterior * 100, 2), "%")
```
