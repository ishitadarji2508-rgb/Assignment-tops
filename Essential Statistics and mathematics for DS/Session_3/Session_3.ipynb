1. Given an array of daily step counts from your fitness app for one week (e.g., [4500, 7000, 5000, 8000, 4000, 9000, 6000]), calculate the variance by hand using the variance formula and show each calculation step.


```python
import numpy as np

steps = np.array([4500, 7000, 5000, 8000, 4000, 9000, 6000])

mean = np.mean(steps)
print("Mean:", round(mean, 2))

print("\nStep-by-step calculations:")
squared_diff_sum = 0

for step in steps:
    deviation = step - mean
    squared_deviation = deviation ** 2
    squared_diff_sum += squared_deviation
    print(f"Step Count = {step}, Deviation = {deviation:.2f}, Squared Deviation = {squared_deviation:.2f}")

variance = squared_diff_sum / (len(steps) - 1)

print("\nSum of Squared Deviations:", round(squared_diff_sum, 2))
print("Sample Variance:", round(variance, 2))
```

2. Write a Python function called calculate_standard_deviation(scores) that takes a list of exam scores and returns the standard deviation rounded to two decimal places.


```python
import math

def calculate_standard_deviation(scores):
    n = len(scores)
    mean = sum(scores) / n

    variance = sum((x - mean) ** 2 for x in scores) / (n - 1)

    std_dev = math.sqrt(variance)

    return round(std_dev, 2)

exam_scores = [75, 80, 85, 90, 95, 88, 92]

result = calculate_standard_deviation(exam_scores)
print("Standard Deviation:", result)
```

3. Imagine two friends track their daily spending on Zomato over 5 days. Friend A spends [200, 200, 200, 200, 200] and Friend B spends [100, 300, 150, 400, 50]. Calculate the standard deviation for both and explain in 2-3 lines who has more consistent spending and why.


```python
import statistics

friend_A = [200, 200, 200, 200, 200]
friend_B = [100, 300, 150, 400, 50]

std_A = statistics.stdev(friend_A)
std_B = statistics.stdev(friend_B)

print("Friend A Standard Deviation:", round(std_A, 2))
print("Friend B Standard Deviation:", round(std_B, 2))

if std_A < std_B:
    print("\nFriend A has more consistent spending because the standard deviation is lower.")
else:
    print("\nFriend B has more consistent spending because the standard deviation is lower.")
```

4. You are given the monthly salary data of 6 employees in a startup: [28000, 29000, 27000, 60000, 26500, 27500]. Calculate the variance and standard deviation, then interpret in 2-3 lines what the high or low spread means for this team.

Var(X) = E(X^2) - [E(X)]^2

```python
import numpy as np

salaries = np.array([28000, 29000, 27000, 60000, 26500, 27500])

variance = np.var(salaries)
std_deviation = np.std(salaries)

print("Variance:", round(variance, 2))
print("Standard Deviation:", round(std_deviation, 2))
```
