1. Generate 100 random numbers in Python that follow a normal distribution with mean 50 and standard deviation 10 using numpy, then plot the bell curve using matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

mean = 50
std_dev = 10
data = np.random.normal(mean, std_dev, 100)

plt.figure(figsize=(8, 5))
plt.hist(data, bins=10, density=True, alpha=0.7, edgecolor='black')

plt.title("Normal Distribution (Mean = 50, Std Dev = 10)")
plt.xlabel("Value")
plt.ylabel("Density")

plt.show()
```

2. Given an array of 20 daily step counts from your fitness tracker app, calculate the mean, median, and mode, and check if they are approximately equal as expected in a normal distribution.<br><br><em><strong>Hint:</strong> Use numpy.mean(), numpy.median(), and scipy.stats.mode().</em>


```python
import numpy as np
from scipy import stats

steps = np.array([
    8000, 8500, 9000, 9500, 10000,
    9200, 8800, 9700, 10100, 9800,
    9400, 9100, 8900, 9600, 9300,
    9000, 9500, 9800, 10200, 8700
])

mean = np.mean(steps)

median = np.median(steps)

mode_result = stats.mode(steps, keepdims=True)
mode = mode_result.mode[0]

print("Daily Step Counts:", steps)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)

if abs(mean - median) < 500 and abs(mean - mode) < 500:
    print("The mean, median, and mode are approximately equal.")
else:
    print("The mean, median, and mode are not approximately equal.")
```

3. Simulate the scores of 200 players in an online game (e.g., PUBG, Free Fire) using a normal distribution with mean 70 and standard deviation 15, then count how many players scored within one standard deviation (between 55 and 85) of the mean.<br><br><em><strong>Hint:</strong> Use the 68-95-99 rule to check your answer.</em>


```python
import numpy as np

np.random.seed(42)

mean = 70
std_dev = 15
scores = np.random.normal(mean, std_dev, 200)

lower_limit = mean - std_dev   # 55
upper_limit = mean + std_dev   # 85

count = np.sum((scores >= lower_limit) & (scores <= upper_limit))

print("Total Players:", len(scores))
print("Players scoring between 55 and 85:", count)
print("Percentage:", round((count / len(scores)) * 100, 2), "%")

expected = int(0.68 * len(scores))
print("Expected (approximately 68%):", expected, "players")
```

4. Pick any real-world scenario (like delivery times for Zomato orders or playlist song durations on Spotify) and explain in 3-4 lines why the normal distribution might be a good fit for modeling this data.

Scenario: Zomato Delivery Times

Zomato delivery times often follow a normal distribution because most orders are delivered around the average time, while only a few are delivered much earlier or much later. Small factors such as traffic, weather, and restaurant preparation time create natural variations around the average. As a result, the data forms a bell-shaped curve, making the normal distribution a good model for predicting delivery times.
