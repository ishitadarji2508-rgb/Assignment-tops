2. Take the following list of Instagram influencer follower counts (in thousands): [5, 7, 8, 8, 9, 10, 12, 15, 95]. Calculate the mean and median, then state which is higher and what that tells you about the skewness of the data.<br><br><em><strong>Hint:</strong> If the mean is pulled away from the median, the data is skewed in that direction.</em>


```python
import numpy as np

followers = [5, 7, 8, 8, 9, 10, 12, 15, 95]

mean = np.mean(followers)
median = np.median(followers)

print("Mean:", mean)
print("Median:", median)

if mean > median:
    print("The mean is higher than the median.")
    print("The data is right (positively) skewed.")
elif mean < median:
    print("The mean is lower than the median.")
    print("The data is left (negatively) skewed.")
else:
    print("The mean and median are equal.")
    print("The data is approximately symmetric.")
```

**Example: YouTube Video Views**

A good example of **right-skewed data** is the number of views on YouTube videos. Most videos receive relatively **few views** (such as hundreds or thousands), while a small number of viral videos receive **millions or even billions of views**. These very large values create a long tail on the **right side** of the distribution, making the data **right skewed**.


4. Given this dataset of Flipkart product ratings: [3, 3, 4, 4, 4, 5, 5, 5, 5, 5], draw a rough graph and state if the distribution is symmetrical, left skewed, or right skewed. Justify your answer in one line.

3. Find a real-world example of right-skewed data from any app you use (like Zomato order amounts, Paytm wallet balances, or YouTube video views). Briefly describe the dataset and explain why you think it is right skewed.


```python
import matplotlib.pyplot as plt

ratings = [3, 3, 4, 4, 4, 5, 5, 5, 5, 5]

plt.figure(figsize=(6, 4))
plt.hist(ratings, bins=[2.5, 3.5, 4.5, 5.5], edgecolor='black')

plt.title("Distribution of Flipkart Product Ratings")
plt.xlabel("Ratings")
plt.ylabel("Frequency")
plt.xticks([3, 4, 5])

plt.grid(axis='y')
plt.show()

print("Distribution: Left-skewed (Negatively Skewed)")
print("Reason: Most ratings are 5, with fewer 4s and 3s, so the tail extends to the left.")
```

Distribution: Left-skewed (negatively skewed)

Justification: Most ratings are 5, with fewer 4s and 3s, so the tail extends toward the lower ratings (left side), indicating a left-skewed distribution.
