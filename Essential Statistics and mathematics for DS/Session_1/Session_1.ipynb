4. Suppose you are building a new playlist recommendation feature for a Spotify-like app. List 3 basic statistical terms (such as mean, median, mode, variance, etc.) that would be useful in analyzing user listening data, and explain each with a one-line example.

For a playlist recommendation system in a Spotify-like app, these statistical terms are very useful:

Mean (Average)
Example: If a user listens to songs for 2, 4, and 6 minutes, the mean listening time is 4 minutes, helping estimate typical engagement per track.
Median
Example: If play counts for songs are [1, 3, 100], the median is 3, which helps avoid distortion from one highly viral song.
Mode
Example: If a user mostly listens to “Pop” genre songs, then Pop is the mode, helping the system recommend more Pop playlists.
