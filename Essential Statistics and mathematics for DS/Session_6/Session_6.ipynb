2. Given the data: Group A (users shown a new Zomato home page) had an average order value of ₹350 from 30 users, and Group B (old home page) had an average order value of ₹320 from 30 users. Assume both groups have a standard deviation of ₹50. Calculate the p-value using a two-sample t-test (you may use Python, R, or an online calculator), and decide whether to accept or reject the null hypothesis at a 0.05 significance level.<br><br><em><strong>Hint:</strong> Use scipy.stats.ttest_ind in Python or any online t-test calculator if you are not comfortable with code.</em>


```python
from scipy import stats
import numpy as np

np.random.seed(42)

group_A = np.random.normal(loc=350, scale=50, size=30)
group_B = np.random.normal(loc=320, scale=50, size=30)

t_stat, p_value = stats.ttest_ind(group_A, group_B)

print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 4))

alpha = 0.05

if p_value < alpha:
    print("Reject the Null Hypothesis (H0)")
else:
    print("Fail to Reject the Null Hypothesis (H0)")
```

The new "Add to Wishlist" button received 520 clicks, while the old button received 480 clicks, a difference of 40 clicks out of 1000 users. This difference may indicate that the new design performs better, but we cannot conclude that it is statistically significant without performing a statistical test.

The p-value helps determine whether the observed difference is likely due to chance. If the p-value is less than 0.05, we reject the null hypothesis and conclude that the new button likely improves user clicks. If the p-value is greater than 0.05, we fail to reject the null hypothesis, meaning the difference could simply be due to random variation.
