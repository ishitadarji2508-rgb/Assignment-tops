2. Build a small Python script that fetches the latest price of Bitcoin and Ethereum from the CoinGecko API and displays them with the current date and time.


```python
import requests
from datetime import datetime
url = "https://api.coingecko.com/api/v3/simple/price"

params = {
    "ids": "bitcoin,ethereum",
    "vs_currencies": "usd"
}

response = requests.get(url, params=params)
data = response.json()

now = datetime.now().strftime("%d-%m-%Y %H:%M:%S")

btc_price = data["bitcoin"]["usd"]
eth_price = data["ethereum"]["usd"]

print("Time:", now)
print("Bitcoin Price: $", btc_price)
print("Ethereum Price: $", eth_price)
```

3. Use the NASA Astronomy Picture of the Day (APOD) API to get today's image title and explanation, and save the image to your local system.<br><br><em><strong>Hint:</strong> The API returns a URL for the image — use requests to download it.</em>

```python
import requests

API_KEY = "DEMO_KEY"

url = "https://api.nasa.gov/planetary/apod"

params = {
    "api_key": API_KEY
}

response = requests.get(url, params=params)
data = response.json()

title = data["title"]
explanation = data["explanation"]
image_url = data["url"]

print(" Title:", title)
print("\n Explanation:\n", explanation)

if data["media_type"] == "image":
    img_data = requests.get(image_url).content

    file_name = "nasa_apod.jpg"

    with open(file_name, "wb") as file:
        file.write(img_data)

    print(f"\n Image saved as {file_name}")
else:
    print("\nToday's APOD is not an image (it may be a video).")
```

4. Create a Python program that fetches the latest COVID-19 case numbers for India from a public COVID API and displays total cases, recovered, and deaths in a table format.


```python
import requests
url = "https://disease.sh/v3/covid-19/countries/India"

response = requests.get(url)
data = response.json()

country = data["country"]
total_cases = data["cases"]
recovered = data["recovered"]
deaths = data["deaths"]
active = data["active"]

print("=" * 50)
print(f"{'COVID-19 Stats':^50}")
print("=" * 50)

print(f"{'Country':<20}: {country}")
print(f"{'Total Cases':<20}: {total_cases}")
print(f"{'Recovered':<20}: {recovered}")
print(f"{'Deaths':<20}: {deaths}")
print(f"{'Active Cases':<20}: {active}")

print("=" * 50)
```

5. Combine data from two APIs: fetch the current temperature in Mumbai from OpenWeatherMap and the latest Bitcoin price from CoinGecko, then display both results together in a single output.<br><br><em><strong>Constraint:</strong> Handle errors gracefully if either API call fails and show an appropriate message.</em>

```python
import requests
from datetime import datetime

WEATHER_API_KEY = "YOUR_API_KEY"
city = "Mumbai"

weather_url = "https://api.openweathermap.org/data/2.5/weather"
weather_params = {
    "q": city,
    "appid": WEATHER_API_KEY,
    "units": "metric"
}

crypto_url = "https://api.coingecko.com/api/v3/simple/price"
crypto_params = {
    "ids": "bitcoin",
    "vs_currencies": "usd"
}

try:
    weather_response = requests.get(weather_url, params=weather_params, timeout=5)
    weather_data = weather_response.json()

    if weather_response.status_code == 200:
        temperature = weather_data["main"]["temp"]
        weather_status = weather_data["weather"][0]["description"]
        weather_result = f"{temperature}°C, {weather_status}"
    else:
        weather_result = f"Weather API error: {weather_data.get('message')}"

except Exception as e:
    weather_result = f"Weather API failed: {str(e)}"

try:
    crypto_response = requests.get(crypto_url, params=crypto_params, timeout=5)
    crypto_data = crypto_response.json()

    if crypto_response.status_code == 200:
        btc_price = crypto_data["bitcoin"]["usd"]
        crypto_result = f"${btc_price}"
    else:
        crypto_result = "Crypto API error"

except Exception as e:
    crypto_result = f"Crypto API failed: {str(e)}"

now = datetime.now().strftime("%d-%m-%Y %H:%M:%S")

print("=" * 50)
print(f"{'LIVE DATA DASHBOARD':^50}")
print("=" * 50)
print(f"Time           : {now}")
print(f"City           : {city}")
print(f"Temperature    : {weather_result}")
print(f"Bitcoin Price  : {crypto_result}")
print("=" * 50)
```
