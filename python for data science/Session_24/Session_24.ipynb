1. Use the requests library to fetch the top 10 cryptocurrencies and their current prices in USD from the CoinGecko API (https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=10&page=1). Print the name and price of each coin.


```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1
}

try:
    response = requests.get(url, params=params, timeout=5)
    data = response.json()

    if response.status_code == 200:
        print("=" * 50)
        print(f"{'TOP 10 CRYPTOCURRENCIES':^50}")
        print("=" * 50)

        for coin in data:
            name = coin["name"]
            price = coin["current_price"]
            print(f"{name:<15} : ${price}")

        print("=" * 50)

    else:
        print("API Error:", data.get("error", "Unknown error"))

except Exception as e:
    print("Request failed:", str(e))
```

2. Extend your script to also fetch and display the 24-hour price change percentage, 24-hour high, and 24-hour low for each of the top 10 cryptocurrencies.


```python
import requests

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1,
    "price_change_percentage": "24h"
}

try:
    response = requests.get(url, params=params, timeout=5)
    data = response.json()

    if response.status_code == 200:
        print("=" * 80)
        print(f"{'TOP 10 CRYPTOCURRENCIES (24H DATA)':^80}")
        print("=" * 80)

        for coin in data:
            name = coin.get("name", "N/A")
            price = coin.get("current_price", 0)
            change = coin.get("price_change_percentage_24h", 0)
            high = coin.get("high_24h", 0)
            low = coin.get("low_24h", 0)

            print(f"\n{name}")
            print(f"Price (USD)   : ${price}")
            print(f"24h Change    : {change:.2f}%")
            print(f"24h High      : ${high}")
            print(f"24h Low       : ${low}")

        print("=" * 80)

    else:
        print("API Error:", data.get("error", "Unknown error"))

except Exception as e:
    print("Request failed:", str(e))
```

3. Save the fetched data (name, current price, price change %, 24h high, 24h low) for the top 10 cryptocurrencies into a CSV file named crypto_prices.csv using the csv module.


```python
import requests
import csv
url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1,
    "price_change_percentage": "24h"
}

try:
    response = requests.get(url, params=params, timeout=5)
    data = response.json()

    if response.status_code == 200:

        with open("crypto_prices.csv", "w", newline="", encoding="utf-8") as file:
            writer = csv.writer(file)

            writer.writerow([
                "Name",
                "Current Price (USD)",
                "24h Change (%)",
                "24h High (USD)",
                "24h Low (USD)"
            ])

           
            for coin in data:
                name = coin.get("name")
                price = coin.get("current_price")
                change = coin.get("price_change_percentage_24h")
                high = coin.get("high_24h")
                low = coin.get("low_24h")

                writer.writerow([name, price, change, high, low])

        print(" CSV file 'crypto_prices.csv' created successfully!")

    else:
        print("API Error:", data.get("error", "Unknown error"))

except Exception as e:
    print("Request failed:", str(e))
```

4. Add error handling to your code so that if the API request fails or returns an error, your script prints a user-friendly message instead of crashing.<br><br><em><strong>Hint:</strong> Check the response status code and handle exceptions from the requests library.</em>
Topic Repository URL

```python
import requests
import csv

url = "https://api.coingecko.com/api/v3/coins/markets"

params = {
    "vs_currency": "usd",
    "order": "market_cap_desc",
    "per_page": 10,
    "page": 1,
    "price_change_percentage": "24h"
}

try:

    response = requests.get(url, params=params, timeout=5)

    if response.status_code != 200:
        print(" Failed to fetch data from CoinGecko API.")
        print("Status Code:", response.status_code)
        print("Message:", response.text)
        exit()

    data = response.json()

    with open("crypto_prices.csv", "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)

        writer.writerow([
            "Name",
            "Current Price (USD)",
            "24h Change (%)",
            "24h High (USD)",
            "24h Low (USD)"
        ])

    
        for coin in data:
            writer.writerow([
                coin.get("name", "N/A"),
                coin.get("current_price", "N/A"),
                coin.get("price_change_percentage_24h", "N/A"),
                coin.get("high_24h", "N/A"),
                coin.get("low_24h", "N/A")
            ])

    print("Data saved successfully to 'crypto_prices.csv'")

except requests.exceptions.Timeout:
    print("Request timed out. Please try again later.")

except requests.exceptions.ConnectionError:
    print("Network error. Check your internet connection.")

except Exception as e:
    print("Something went wrong:", str(e))
```
