1. Create a Python class called Playlist with a private attribute _songs (a list) and a public method add_song(song) to add a song title to the playlist. Print the playlist after adding 3 songs.


```python
class Playlist:
    def __init__(self):
        self._songs = []   

    def add_song(self, song):
        self._songs.append(song)

    def show_playlist(self):
        print("Playlist Songs:")
        for song in self._songs:
            print("-", song)



my_playlist = Playlist()

my_playlist.add_song("Blinding Lights")
my_playlist.add_song("Shape of You")
my_playlist.add_song("Believer")

my_playlist.show_playlist()
```

2. Build a Product class for a Flipkart-style app with a private attribute _price. Implement get_price() and set_price() methods to access and update the price. Demonstrate setting and getting the price for a product object.


```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self._price = price 

    def get_price(self):
        return self._price

    def set_price(self, new_price):
        if new_price > 0:
            self._price = new_price
        else:
            print("Price must be greater than 0")

product = Product("Smartphone", 15000)

print("Original Price:", product.get_price())

product.set_price(17000)

print("Updated Price:", product.get_price())
```

3. Create a Movie class with a private attribute _rating (float between 0 and 10). Write getter and setter methods for _rating. The setter should only allow values between 0 and 10; if an invalid value is given, print an error message.<br><br><em><strong>Constraint:</strong> Do not allow direct access to _rating outside the class.</em>


```python
class Movie:
    def __init__(self, title, rating):
        self.title = title
        self._rating = None
        self.set_rating(rating)

    def get_rating(self):
        return self._rating

    def set_rating(self, new_rating):
        if 0 <= new_rating <= 10:
            self._rating = new_rating
        else:
            print("Error: Rating must be between 0 and 10")

movie = Movie("Inception", 8.8)

print("Original Rating:", movie.get_rating())

movie.set_rating(9.5)
print("Updated Rating:", movie.get_rating())

movie.set_rating(15)
print("Final Rating:", movie.get_rating())
```

4. Design a simple abstract class PaymentMethod with an abstract method pay(amount). Then, create two subclasses: Paytm and PhonePe, each implementing pay(amount) to print a different message. Instantiate both and call their pay methods with any amount.

```python
from abc import ABC, abstractmethod

class PaymentMethod(ABC):

    @abstractmethod
    def pay(self, amount):
        pass

class Paytm(PaymentMethod):
    def pay(self, amount):
        print(f"Payment of Rs.{amount} done using Paytm")


class PhonePe(PaymentMethod):
    def pay(self, amount):
        print(f"Payment of Rs.{amount} done using PhonePe")


paytm = Paytm()
phonepe = PhonePe()

paytm.pay(500)
phonepe.pay(750)
```
