3. Simulate an async data fetch from two different APIs (for example, fetch trending songs from Spotify and trending movies from BookMyShow) using Python's asyncio and httpx library, and print both results when done.<br><br><em><strong>Hint:</strong> Use asyncio.gather() to run both requests concurrently.</em>


```python
import asyncio
import httpx

SPOTIFY_URL = "https://jsonplaceholder.typicode.com/posts/1"
BOOKMYSHOW_URL = "https://jsonplaceholder.typicode.com/posts/2"

async def fetch_data(client, url):
    response = await client.get(url)
    return response.json()

async def main():
    async with httpx.AsyncClient() as client:

        spotify_task = fetch_data(client, SPOTIFY_URL)
        movie_task = fetch_data(client, BOOKMYSHOW_URL)

        spotify_data, movie_data = await asyncio.gather(
            spotify_task,
            movie_task
        )

        print(" Trending Songs Data (Spotify mock):")
        print(spotify_data)

        print("\nTrending Movies Data (BookMyShow mock):")
        print(movie_data)
await main()
```

4. Many APIs require Bearer tokens for authentication. Write a function get_user_profile() that calls a mock API endpoint (e.g., https://jsonplaceholder.typicode.com/users/1) using a fake Bearer token in the Authorization header, and prints the user's name.<br><br><em><strong>Constraint:</strong> Use the 'Authorization: Bearer <token>' header format.</em>


```python
import requests

def get_user_profile():
    url = "https://jsonplaceholder.typicode.com/users/1"
    
    token = "fake_token_12345"

    headers = {
        "Authorization": f"Bearer {token}"
    }

    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        data = response.json()
        print("User Name:", data["name"])
    else:
        print("Request failed with status:", response.status_code)


get_user_profile()
```

5. Use ChatGPT or Copilot to generate Python code that demonstrates the first step of an OAuth 2.0 login flow (for example, generating the URL to redirect a user to Spotify's OAuth login page). Paste the generated code and briefly explain what it does.

```python
import urllib.parse

def generate_spotify_auth_url(client_id, redirect_uri, scopes):
    base_url = "https://accounts.spotify.com/authorize"

    params = {
        "response_type": "code",
        "client_id": client_id,
        "redirect_uri": redirect_uri,
        "scope": " ".join(scopes),
    }

    url = f"{base_url}?{urllib.parse.urlencode(params)}"
    return url

client_id = "your_spotify_client_id"
redirect_uri = "https://yourapp.com/callback"
scopes = ["user-read-email", "user-read-private"]

auth_url = generate_spotify_auth_url(client_id, redirect_uri, scopes)

print("Open this URL in browser to login:\n")
print(auth_url)
```
