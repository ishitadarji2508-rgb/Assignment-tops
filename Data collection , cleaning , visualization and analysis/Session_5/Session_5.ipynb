2. Use pd.read_sql_query() to fetch only the 'name' and 'rating' columns from a 'movies' table (think BookMyShow-like data) where rating is above 8, and print the resulting DataFrame.<br><br><em><strong>Hint:</strong> Write a custom SQL SELECT query as the first argument to pd.read_sql_query().</em>


```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine(
    "mysql+pymysql://root:123456789@localhost:3306/movies_db"
)

query = """
SELECT name, rating
FROM movies
WHERE rating > 8
"""

df = pd.read_sql_query(query, engine)

print(df)
```

4. You have two CSV files: 'orders.csv' (order_id, user_id, amount) and 'users.csv' (user_id, username). Load both into DataFrames using pathlib for file paths, then merge them on 'user_id' to show a combined table with username and amount.


```python
import pandas as pd
from pathlib import Path

orders_file = Path("orders.csv")
users_file = Path("users.csv")

orders_df = pd.read_csv(orders_file)
users_df = pd.read_csv(users_file)

merged_df = pd.merge(
    orders_df,
    users_df,
    on="user_id"
)

print(merged_df[["username", "amount"]])
```

5. Concatenate two DataFrames representing 'today_orders' and 'yesterday_orders' (each with columns: order_id, item, price), and display the combined DataFrame.<br><br><em><strong>Constraint:</strong> Use pd.concat() and reset the index after concatenation.</em>

```python
import pandas as pd

today_orders = pd.DataFrame({
    "order_id": [101, 102],
    "item": ["Pizza", "Burger"],
    "price": [250, 150]
})

yesterday_orders = pd.DataFrame({
    "order_id": [99, 100],
    "item": ["Pasta", "Sandwich"],
    "price": [200, 120]
})

combined_df = pd.concat(
    [today_orders, yesterday_orders],
    ignore_index=True
).reset_index(drop=True)

print(combined_df)
```
