4. Download a large Excel file (at least 10,000 rows) of Flipkart product listings. Use pd.read_excel() with the chunksize parameter to read the file in chunks of 2000 rows, and print the number of rows in each chunk as you iterate.


```python
import pandas as pd

n_rows = 10000

df = pd.DataFrame({
    "ProductID": range(1, n_rows + 1),
    "ProductName": [f"Product {i}" for i in range(1, n_rows + 1)],
    "Category": ["Electronics"] * n_rows,
    "Price": [(i % 500) + 100 for i in range(1, n_rows + 1)],
    "Rating": [3.5 + (i % 15) / 10 for i in range(1, n_rows + 1)]
})

df.to_excel("flipkart_products.xlsx", index=False)

print("Created flipkart_products.xlsx with", len(df), "rows")
```

5. Use ChatGPT or Copilot to generate a Python code snippet that reads a semicolon-separated CSV of Paytm transactions, detects missing values, and prints out which columns have nulls. Test the code with a small sample file and fix any errors you encounter.

```python
import pandas as pd

df = pd.read_csv("paytm_transactions.csv", sep=";")

null_counts = df.isnull().sum()

columns_with_nulls = null_counts[null_counts > 0]

if len(columns_with_nulls) == 0:
    print("No missing values found.")
else:
    print("Columns with missing values:")
    for col, count in columns_with_nulls.items():
        print(f"{col}: {count} missing value(s)")
```
