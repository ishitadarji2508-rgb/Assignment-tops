1. Use plt.subplots() to create a 2x2 grid of subplots and plot four different types of charts (line, bar, scatter, and pie) using any sample data of your choice.


```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
values = [10, 25, 15, 30, 20]

categories = ["A", "B", "C", "D"]
counts = [40, 25, 20, 15]

x = [1, 2, 3, 4, 5]
y = [5, 15, 10, 20, 25]

pie_labels = ["Chrome", "Safari", "Edge", "Others"]
pie_sizes = [55, 25, 10, 10]

fig, axes = plt.subplots(2, 2, figsize=(10, 8))

axes[0, 0].plot(days, values, marker='o')
axes[0, 0].set_title("Line Plot")
axes[0, 0].set_xlabel("Days")
axes[0, 0].set_ylabel("Values")


axes[0, 1].bar(categories, counts)
axes[0, 1].set_title("Bar Chart")
axes[0, 1].set_xlabel("Category")
axes[0, 1].set_ylabel("Count")

axes[1, 0].scatter(x, y)
axes[1, 0].set_title("Scatter Plot")
axes[1, 0].set_xlabel("X")
axes[1, 0].set_ylabel("Y")


axes[1, 1].pie(pie_sizes, labels=pie_labels, autopct='%1.1f%%')
axes[1, 1].set_title("Pie Chart")

plt.tight_layout()
plt.show()
```

2. Plot a comparison of average delivery times for Zomato, Swiggy, and Domino's using a bar chart, and style each bar with a different color, linestyle, and linewidth using Matplotlib plot styling options.


```python
import matplotlib.pyplot as plt

platforms = ["Zomato", "Swiggy", "Domino's"]
avg_times = [32, 28, 40]

colors = ["orange", "green", "red"]
edge_linestyles = ["--", "-.", ":"]
edge_linewidths = [2, 3, 4]

plt.figure(figsize=(8, 5))

bars = plt.bar(platforms, avg_times, color=colors)

for bar, ls, lw in zip(bars, edge_linestyles, edge_linewidths):
    bar.set_edgecolor("black")
    bar.set_linestyle(ls)
    bar.set_linewidth(lw)

plt.title("Average Delivery Time Comparison")
plt.xlabel("Food Delivery Platforms")
plt.ylabel("Average Delivery Time (minutes)")

for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,
             f"{height}", ha='center')

plt.tight_layout()
plt.show()
```

3. Create a multi-axis chart that shows the number of Instagram followers (left y-axis) and average daily posts (right y-axis) for five influencers. Use different colors and markers for each axis.


```python
import matplotlib.pyplot as plt

influencers = ["Influencer A", "Influencer B", "Influencer C", "Influencer D", "Influencer E"]

followers = [1200000, 950000, 1800000, 600000, 2200000]   #
daily_posts = [2, 5, 3, 4, 1]            
fig, ax1 = plt.subplots(figsize=(10, 6))

color1 = "blue"
ax1.set_xlabel("Influencers")
ax1.set_ylabel("Followers", color=color1)
ax1.plot(influencers, followers, color=color1, marker="o", linewidth=2, label="Followers")
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()
color2 = "green"
ax2.set_ylabel("Avg Daily Posts", color=color2)
ax2.plot(influencers, daily_posts, color=color2, marker="s", linestyle="--", linewidth=2, label="Daily Posts")
ax2.tick_params(axis='y', labelcolor=color2)

plt.title("Instagram Followers vs Average Daily Posts")

fig.tight_layout()

plt.show()
```

4. Plot the number of tickets sold for five recent Bollywood movies (categorical) versus their IMDB ratings (numeric) using a scatter plot. Add annotations to display the movie names above each point.


```python
import matplotlib.pyplot as plt

movies = ["Pathaan", "Jawan", "Animal", "Dunki", "Fighter"]
imdb_ratings = [6.5, 7.0, 6.8, 7.2, 6.9]
tickets_sold = [50000000, 60000000, 45000000, 40000000, 35000000]

plt.figure(figsize=(8, 6))

plt.scatter(imdb_ratings, tickets_sold, color="purple", marker="o", s=100)

for i in range(len(movies)):
    plt.annotate(
        movies[i],
        (imdb_ratings[i], tickets_sold[i]),
        textcoords="offset points",
        xytext=(0, 10),
        ha='center'
    )

plt.title("IMDb Ratings vs Tickets Sold (Bollywood Movies)")
plt.xlabel("IMDb Rating")
plt.ylabel("Tickets Sold")

plt.grid(True, linestyle="--", alpha=0.5)

plt.show()
```

5. Add a custom text annotation to a Matplotlib chart showing Flipkart's monthly sales, marking the highest sales point with the label 'Big Billion Days'.<br><br><em><strong>Hint:</strong> Use the ax.annotate() function to place the label at the correct data point.</em>

```python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct"]
sales = [120, 150, 170, 160, 200, 250, 230, 220, 210, 400]

fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(months, sales, marker="o", linewidth=2, color="blue")

max_index = sales.index(max(sales))
max_month = months[max_index]
max_value = sales[max_index]

ax.annotate(
    "Big Billion Days",
    xy=(max_month, max_value),         
    xytext=(max_month, max_value + 30), 
    ha="center",
    fontsize=10,
    color="red",
    arrowprops=dict(arrowstyle="->", color="red", linewidth=1.5)
)

ax.set_title("Flipkart Monthly Sales")
ax.set_xlabel("Months")
ax.set_ylabel("Sales (Units)")

plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
