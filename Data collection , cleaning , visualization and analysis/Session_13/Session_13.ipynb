1. Load the 'tips' dataset from Seaborn and create a pairplot to quickly visualize relationships between all numeric variables.


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.pairplot(
    tips,
    hue="sex",        
    diag_kind="hist"    
)

plt.show()
```

2. Using the 'flights' dataset from Seaborn, generate a heatmap showing the correlation between months and years based on the number of passengers.<br><br><em><strong>Hint:</strong> Use pivot_table to reshape the data before plotting the heatmap.</em>


```python
import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset("flights")

flight_pivot = flights.pivot_table(
    index="month",
    columns="year",
    values="passengers"
)

plt.figure(figsize=(12, 6))
sns.heatmap(flight_pivot, cmap="YlOrRd", annot=True, linewidths=0.5)

plt.title("Flights Passenger Heatmap (Month vs Year)")
plt.xlabel("Year")
plt.ylabel("Month")

plt.show()
```

3. Create a relplot using the 'fmri' dataset from Seaborn to visualize how the signal changes over time for different event types.


```python
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

sns.relplot(
    data=fmri,
    x="timepoint",
    y="signal",
    hue="event",     
    kind="line",
    palette="Set2",
    height=5,
    aspect=1.5
)

plt.title("FMRI Signal Over Time by Event Type")

plt.show()
```

4. Pick any categorical column from the 'titanic' dataset (like 'class' or 'sex') and use catplot to display the survival rate for each group with confidence intervals.


```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

sns.catplot(
    data=titanic,
    x="class",
    y="survived",
    kind="bar",
    palette="Set2",
    errorbar=("ci", 95)  
)

plt.title("Survival Rate by Passenger Class (Titanic Dataset)")

plt.show()
```

5. Use jointplot on the 'penguins' dataset to visualize the relationship between bill_length_mm and flipper_length_mm, and fit a regression line on the scatter plot.<br><br><em><strong>Hint:</strong> Set kind='reg' in jointplot for regression line.</em>

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

penguins = penguins.dropna(subset=["bill_length_mm", "flipper_length_mm"])

sns.jointplot(
    data=penguins,
    x="bill_length_mm",
    y="flipper_length_mm",
    kind="reg",        
    height=6,
    scatter_kws={"alpha": 0.6},
    line_kws={"color": "red"}
)

plt.suptitle("Penguins: Bill Length vs Flipper Length (Regression)", y=1.02)

plt.show()
```
