1. Given a list of delivery timestamps as strings (e.g., ['2024-06-01 14:30', '2024-06-02 09:15', '2024-06-03 20:45']), use pandas and pd.to_datetime() to convert them into datetime objects and print the result.


```python
import pandas as pd

timestamps = ['2024-06-01 14:30', '2024-06-02 09:15', '2024-06-03 20:45']

dt_series = pd.to_datetime(timestamps)

print("Converted datetime objects:")
print(dt_series)
```

2. Load a CSV file containing order dates from a Flipkart-style order history (column: 'order_date', format: 'YYYY-MM-DD HH:MM:SS'). Extract the year, month, and weekday for each order and add them as new columns in the DataFrame.


```python
import pandas as pd

df = pd.read_csv("flipkart_orders.csv")

df["order_date"] = pd.to_datetime(df["order_date"])

df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["weekday"] = df["order_date"].dt.day_name()

print("Updated DataFrame:")
print(df)
```

3. Set the 'order_date' column as the index of your DataFrame and use resampling to calculate the total number of orders placed each week.<br><br><em><strong>Hint:</strong> Use df.resample('W').size() after setting the datetime index.</em>


```python
import pandas as pd

df = pd.DataFrame({
    "order_id": [1, 2, 3, 4, 5, 6],
    "order_date": [
        "2024-06-01 10:15:30",
        "2024-06-02 14:45:10",
        "2024-06-03 09:20:05",
        "2024-06-08 21:10:50",
        "2024-06-09 16:05:15",
        "2024-06-10 11:30:00"
    ]
})


df["order_date"] = pd.to_datetime(df["order_date"])

df = df.set_index("order_date")

weekly_orders = df.resample("W").size()

print("Weekly order counts:")
print(weekly_orders)
```

4. Suppose you have a DataFrame of Instagram posts with a 'posted_at' column in UTC. Convert these timestamps to 'Asia/Kolkata' timezone and display the first 5 converted times.


```python
import pandas as pd


df = pd.DataFrame({
    "post_id": [1, 2, 3, 4, 5, 6],
    "posted_at": [
        "2024-06-01 10:00:00",
        "2024-06-01 14:30:00",
        "2024-06-02 18:45:00",
        "2024-06-03 02:15:00",
        "2024-06-03 22:50:00",
        "2024-06-04 06:10:00"
    ]
})

df["posted_at"] = pd.to_datetime(df["posted_at"]).dt.tz_localize("UTC")

df["posted_at"] = df["posted_at"].dt.tz_convert("Asia/Kolkata")

print(df["posted_at"].head(5))
```

5. Create a new feature called 'is_weekend' in your DataFrame that marks True if an order was placed on Saturday or Sunday, and False otherwise.<br><br><em><strong>Constraint:</strong> Do not use any external libraries except pandas and numpy.</em>

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "order_id": [1, 2, 3, 4, 5, 6],
    "order_date": [
        "2024-06-01 10:15:30", 
        "2024-06-02 14:45:10",  
        "2024-06-03 09:20:05", 
        "2024-06-04 21:10:50",  
        "2024-06-08 16:05:15", 
        "2024-06-09 11:30:00" 
    ]
})

df["order_date"] = pd.to_datetime(df["order_date"])

# Create is_weekend feature (Saturday=5, Sunday=6)
df["is_weekend"] = df["order_date"].dt.weekday.isin([5, 6])

print(df)
```
