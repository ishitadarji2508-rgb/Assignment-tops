
1. Download a sample CSV file of IPL player stats (include columns like player name, runs, matches, and some missing values). Load it into a pandas DataFrame and use isnull() and notnull() to print how many missing values are present in each column.


```python
import pandas as pd

df = pd.read_csv("ipl_player_stats.csv")

print("DataFrame:")
print(df)

print("\nMissing values in each column:")
print(df.isnull().sum())

print("\nNon-missing values in each column:")
print(df.notnull().sum())
```

2. Using the loaded IPL player stats DataFrame, apply dropna(axis=0, how='any') to remove all rows with any missing data and display the shape of the DataFrame before and after dropping.


```python
import pandas as pd

df = pd.read_csv("ipl_player_stats.csv")

print("Shape before dropna:", df.shape)

df_clean = df.dropna(axis=0, how='any')

print("Shape after dropna:", df_clean.shape)

print("\nCleaned DataFrame:")
print(df_clean)
```

3. For the 'runs' column in your IPL player stats DataFrame, use fillna() to replace missing values with the mean of the column. Print the updated column to verify the changes.


```python
import pandas as pd

df = pd.read_csv("ipl_player_stats.csv")

runs_mean = df["runs"].mean()

df["runs"] = df["runs"].fillna(runs_mean)

print("Updated 'runs' column:")
print(df["runs"])
```

4. Simulate a Zomato-style restaurant ratings dataset with some missing ratings. Use forward fill (method='ffill') to fill missing values in the ratings column, then use backward fill (method='bfill') for any remaining missing values. Show the before and after results.

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "restaurant": ["Pizza Hub", "Burger Point", "Spice Villa",
                   "Food Corner", "Cafe Delight", "Tandoori House"],
    "rating": [np.nan, 4.2, np.nan, 4.5, np.nan, 4.0]
})

print("Before Filling:")
print(df)

df["rating"] = df["rating"].fillna(method="ffill")
df["rating"] = df["rating"].fillna(method="bfill")

print("\nAfter Filling:")
print(df)
```
