```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
delivery_times = np.random.normal(loc=35, scale=8, size=50)  # mean=35, std=8
delivery_times = np.clip(delivery_times, 10, 60)  # keep values realistic

plt.figure(figsize=(8, 5))
sns.histplot(delivery_times, bins=10, kde=True, color="orange")

plt.title("Distribution of Zomato Delivery Times (50 Orders)")
plt.xlabel("Delivery Time (minutes)")
plt.ylabel("Number of Orders")

plt.grid(True, linestyle="--", alpha=0.4)
plt.show()
```

2. Load the 'tips' dataset from Seaborn and create a boxplot to visualize the distribution of total bill amounts by day of the week, similar to how Zomato might analyze spending patterns across weekdays.


```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

plt.figure(figsize=(8, 5))
sns.boxplot(x="day", y="total_bill", data=tips, palette="Set2")

plt.title("Distribution of Total Bill Amounts by Day")
plt.xlabel("Day of Week")
plt.ylabel("Total Bill")

plt.grid(True, linestyle="--", alpha=0.4)
plt.show()
```

3. Simulate IPL match scores for 8 teams and use a violinplot to compare the run distributions per team. Style your plot using the 'darkgrid' Seaborn theme.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

sns.set_theme(style="darkgrid")

teams = ["MI", "CSK", "RCB", "KKR", "SRH", "RR", "PBKS", "DC"]

np.random.seed(42)

data = []

for team in teams:
    
    runs = np.random.normal(loc=160, scale=25, size=20) 
    runs = np.clip(runs, 80, 260) 
    
    for r in runs:
        data.append([team, r])


df = pd.DataFrame(data, columns=["Team", "Runs"])


plt.figure(figsize=(10, 6))
sns.violinplot(x="Team", y="Runs", data=df, palette="Set2", inner="box")

plt.title("IPL Run Distribution per Team (Simulated Data)")
plt.xlabel("Teams")
plt.ylabel("Runs per Match")

plt.show()
```

4. Create a countplot that displays the number of songs per genre from a list of 40 Spotify tracks, with at least 4 different genres represented.<br><br><em><strong>Hint:</strong> Use a Python list or pandas DataFrame to store your data, then plot with Seaborn.</em>


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import random

sns.set_theme(style="whitegrid")

genres = ["Pop", "Hip-Hop", "Rock", "EDM", "Jazz"]

random.seed(42)
data = {
    "Track": [f"Song {i+1}" for i in range(40)],
    "Genre": [random.choice(genres) for _ in range(40)]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
sns.countplot(x="Genre", data=df, palette="viridis")

plt.title("Number of Songs per Genre (40 Spotify Tracks)")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")

plt.show()
```

5. Use Seaborn's kdeplot to visualize the distribution of daily step counts for a week, as if analyzing data from a fitness app. Apply the 'whitegrid' theme and customize the plot color.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")

steps = [5200, 7800, 6500, 9000, 11000, 7200, 8400]

plt.figure(figsize=(8, 5))
sns.kdeplot(
    steps,
    fill=True,
    color="teal",
    linewidth=2
)

plt.title("Distribution of Daily Step Counts (Weekly Fitness Data)")
plt.xlabel("Steps per Day")
plt.ylabel("Density")

plt.show()
```
