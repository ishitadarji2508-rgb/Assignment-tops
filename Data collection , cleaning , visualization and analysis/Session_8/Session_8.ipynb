1. Download a small sample of your recent Zomato order history as a CSV (or create a mock CSV with restaurant names and order dates), then use pandas' duplicated() function to find and print any duplicate orders based on restaurant name and date.


```python
import pandas as pd

df = pd.read_csv("zomato_orders.csv")

print("Full DataFrame:")
print(df)

duplicates = df[df.duplicated(subset=["restaurant", "order_date"], keep=False)]

print("\nDuplicate Orders:")
print(duplicates)
```

2. Given a list of Flipkart product reviews with some duplicate entries, use value_counts() in pandas to identify which review texts are repeated most often and display the top 3 most common duplicate reviews.


```python
import pandas as pd

reviews = [
    "Great product", "Worth the money", "Good quality",
    "Great product", "Excellent", "Worth the money",
    "Not satisfied", "Great product", "Good quality",
    "Average", "Worth the money", "Excellent"
]

df = pd.DataFrame({"review": reviews})

print("All Reviews:")
print(df)

review_counts = df["review"].value_counts()

print("\nTop 3 Most Common Reviews:")
print(review_counts.head(3))
```

3. Create a DataFrame with mock data for Spotify playlists, including playlist names and creator usernames, where some rows are exact duplicates. Use drop_duplicates() to remove duplicate playlists and print the cleaned DataFrame.


```python
import pandas as pd

df = pd.DataFrame({
    "playlist_name": [
        "Top Hits", "Workout Mix", "Chill Vibes",
        "Top Hits", "Bollywood Beats", "Chill Vibes",
        "Focus Flow", "Workout Mix"
    ],
    "creator_username": [
        "user1", "user2", "user3",
        "user1", "user4", "user3",
        "user5", "user2"
    ]
})

print("Before removing duplicates:")
print(df)

df_clean = df.drop_duplicates()

print("\nAfter removing duplicates:")
print(df_clean)
```

4. Suppose you have a DataFrame of Instagram usernames where some entries have typos (like 'insta_queen', 'insta-queen', 'instaqueen'). Use the replace() function to standardize all these variants to 'instaqueen'.


```python
import pandas as pd

df = pd.DataFrame({
    "username": [
        "insta_queen", "insta-queen", "instaqueen",
        "insta_QUEEN", "insta queen", "instaqueen"
    ]
})

print("Before cleaning:")
print(df)

df["username"] = df["username"].replace({
    "insta_queen": "instaqueen",
    "insta-queen": "instaqueen",
    "insta queen": "instaqueen",
    "insta_QUEEN": "instaqueen"
})

print("\nAfter cleaning:")
print(df)
```

5. You have a DataFrame column for payment status from a Paytm-like app with mixed values: 'Yes', 'yes', 'Y', 'No', 'no', 'N', and some with extra spaces. Write code to unify this column so all paid statuses become 1 and all unpaid statuses become 0, trimming whitespace and fixing capitalization where needed.<br><br><em><strong>Hint:</strong> Use str.strip(), str.lower(), and map/replace methods in pandas.</em>

```python
import pandas as pd

df = pd.DataFrame({
    "payment_status": ["Yes", " yes ", "Y", "No", " no", "N", "YES", " y ", "  n "]
})

print("Before cleaning:")
print(df)


cleaned = (
    df["payment_status"]
    .str.strip()        #
    .str.lower()        
)

df["payment_status"] = cleaned.map({
    "yes": 1,
    "y": 1,
    "no": 0,
    "n": 0
})

print("\nAfter cleaning:")
print(df)

print("\nData types:")
print(df.dtypes)
```
