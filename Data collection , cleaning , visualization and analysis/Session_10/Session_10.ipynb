2. Using Matplotlib, create a scatter plot of 10 restaurants from your city with their Zomato ratings on the x-axis and average meal price on the y-axis. Add axis labels and a title to the chart.


```python
import matplotlib.pyplot as plt

restaurants = [
    "Agashiye", "Vishalla", "Swati Snacks", "Tomato's", "La Pino'z",
    "Honest", "Barbeque Nation", "Gordhan Thal", "Karnavati Dabeli",
    "Tinello"
]

ratings = [4.6, 4.4, 4.5, 4.2, 4.1, 4.0, 4.3, 4.5, 4.0, 4.6]

prices = [1200, 1000, 500, 700, 300, 250, 900, 600, 80, 1500]


plt.figure(figsize=(9, 6))
plt.scatter(ratings, prices)

for i, name in enumerate(restaurants):
    plt.text(ratings[i] + 0.01, prices[i] + 10, name, fontsize=8)

plt.title("Zomato Ratings vs Average Meal Price (Ahmedabad Restaurants)")
plt.xlabel("Zomato Rating")
plt.ylabel("Average Meal Price (₹)")

plt.grid(True)
plt.show()
```

3. Build a bar chart that shows the number of orders you or your friends placed on Swiggy, Zomato, and Domino’s in the last month. Use different colors for each bar and add a legend.


```python
import matplotlib.pyplot as plt

platforms = ["Swiggy", "Zomato", "Domino's"]
orders = [12, 8, 5]   # Replace with your actual order counts

colors = ["orange", "red", "blue"]


plt.figure(figsize=(6, 5))
plt.bar(platforms, orders, color=colors, label="Number of Orders")


plt.title("Orders Placed Last Month")
plt.xlabel("Food Delivery Platforms")
plt.ylabel("Number of Orders")

plt.legend()

plt.show()
```

4. Create a histogram of the durations (in minutes) of your last 20 Spotify listening sessions using Matplotlib. Set the number of bins to 5 and customize the color of the bars.


```python
import matplotlib.pyplot as plt

sessions = [15, 22, 30, 18, 45, 27, 35, 40, 12, 25,
            50, 33, 28, 20, 38, 42, 17, 29, 31, 24]

plt.figure(figsize=(8, 5))
plt.hist(sessions, bins=5, color='green', edgecolor='black')

plt.title("Histogram of Spotify Listening Session Durations")
plt.xlabel("Duration (minutes)")
plt.ylabel("Frequency")

plt.show()
```

5. Customize a Matplotlib figure by creating a plot with two subplots: one line plot showing your daily Instagram screen time for a week, and one bar chart showing the number of posts you liked each day. Add appropriate titles, axis labels, and save the figure as social_media_usage.png.<br><br><em><strong>Hint:</strong> Use plt.subplots() to create multiple axes in one figure.</em>

```python
import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

screen_time = [45, 60, 50, 80, 70, 90, 65]


likes = [120, 150, 130, 200, 180, 220, 160]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(days, screen_time, marker='o')
axes[0].set_title("Daily Instagram Screen Time")
axes[0].set_xlabel("Day")
axes[0].set_ylabel("Minutes")

axes[1].bar(days, likes)
axes[1].set_title("Likes per Day")
axes[1].set_xlabel("Day")
axes[1].set_ylabel("Number of Likes")

plt.tight_layout()

plt.show()
```
