2. Create a Sequential model in Keras with three layers: an input Dense layer with 32 units and 'relu' activation, a Dense layer with 16 units and 'relu', and an output Dense layer with 1 unit and 'sigmoid'. Print the model summary.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Dense(32, activation='relu', input_shape=(10,)),
    layers.Dense(16, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.summary()
```

3. uild a simple Keras Functional API model that takes a 10-feature input, passes it through one Dense layer with 20 units and 'relu', then outputs a single value with a Dense layer. Display the model summary and total parameter count.


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

inputs = keras.Input(shape=(10,))

x = layers.Dense(20, activation='relu')(inputs)

outputs = layers.Dense(1)(x)

model = keras.Model(inputs=inputs, outputs=outputs)

model.summary()

print("Total parameters:", model.count_params())
```

5. Use ChatGPT or Copilot to generate Keras code for a model with two inputs (e.g., one for user age and one for number of songs listened on Spotify) and one output (predicting likelihood of buying a premium subscription). Paste the generated code, run it, and note any errors or changes you had to make.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

age_input = keras.Input(shape=(1,), name="age")

songs_input = keras.Input(shape=(1,), name="songs_listened")

x1 = layers.Dense(8, activation="relu")(age_input)

x2 = layers.Dense(8, activation="relu")(songs_input)

combined = layers.concatenate([x1, x2])

x = layers.Dense(16, activation="relu")(combined)

output = layers.Dense(1, activation="sigmoid")(x)

model = keras.Model(inputs=[age_input, songs_input], outputs=output)

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```
