1. Train a simple image classifier on the Fashion MNIST dataset using Keras and the SGD optimizer with default parameters, then record the accuracy and loss after 5 epochs.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

(x_train, y_train), (x_test, y_test) = keras.datasets.fashion_mnist.load_data()

x_train = x_train / 255.0
x_test = x_test / 255.0

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer=keras.optimizers.SGD(),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    x_train,
    y_train,
    epochs=5,
    validation_data=(x_test, y_test),
    verbose=1
)

test_loss, test_acc = model.evaluate(x_test, y_test, verbose=0)

print("\nFinal Test Loss:", test_loss)
print("Final Test Accuracy:", test_acc)
```

3. Experiment with the Adam optimizer on the same Fashion MNIST model and adjust the learning rate to 0.0005; observe and note any changes in training speed or accuracy.<br><br><em><strong>Hint:</strong> In Keras, set learning_rate in the Adam optimizer: Adam(learning_rate=0.0005).</em>

4. Implement a learning rate schedule in your Keras model using the LearningRateScheduler callback to reduce the learning rate by half after every 2 epochs, and train with the Momentum optimizer; report how the learning rate changes during training.

```python
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import LearningRateScheduler

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()

X_train = X_train / 255.0
X_test = X_test / 255.0

model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

initial_lr = 0.01

def lr_schedule(epoch, lr):
    new_lr = initial_lr * (0.5 ** (epoch // 2))
    print(f"Epoch {epoch+1}: Learning Rate = {new_lr}")
    return new_lr

lr_callback = LearningRateScheduler(lr_schedule)

optimizer = SGD(learning_rate=initial_lr, momentum=0.9)

model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history = model.fit(
    X_train,
    y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.2,
    callbacks=[lr_callback]
)
```

Adam is often preferred in industry because it combines the advantages of both SGD with momentum and RMSProp, making it more reliable and easier to use in real-world training without heavy tuning.

Why Adam is commonly preferred

Adam (Adaptive Moment Estimation) improves training in a few important ways:

Adaptive learning rates:
Each parameter gets its own learning rate, so the model adjusts automatically instead of using one fixed value like SGD.
Faster convergence:
It usually reaches good accuracy in fewer epochs, which is important when training large models on expensive hardware.
Handles noisy gradients well:
In real datasets (images, text, user behavior), gradients are often unstable. Adam smooths these updates using moving averages.
Less tuning required:
SGD often needs careful tuning of learning rate and momentum. RMSProp improves this but still needs more manual adjustment compared to Adam.

Because of this, Adam becomes a strong вЂњdefault choiceвЂќ in many deep learning projects.

Comparison intuition
SGD: steady but slow, sensitive to learning rate
RMSProp: adapts learning rate but focuses only on recent gradients
Adam: combines momentum (past trends) + RMSProp (adaptive scaling)

 Impact on user experience:

Key takeaway

Adam is preferred in industry because it is:

Fast to converge
Stable on real noisy data
