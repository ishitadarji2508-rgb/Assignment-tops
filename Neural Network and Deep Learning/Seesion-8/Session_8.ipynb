1. Write a Python function compute_gradient(f, x) that takes a simple function f (like f(x) = x**2) and a value x, and returns the numerical gradient at x using the central difference method.<br><br><em><strong>Hint:</strong> Use (f(x+h) - f(x-h)) / (2*h) with a small h (e.g., 1e-5).</em>


```python
def compute_gradient(f, x, h=1e-5):
    return (f(x + h) - f(x - h)) / (2 * h)

def f(x):
    return x ** 2

x = 3
grad = compute_gradient(f, x)

print("Gradient at x =", x, "is", grad)
```

3. Simulate one step of gradient descent for a function f(w) = (w-5)**2 starting from w=0, with a learning rate of 0.1. Print the updated value of w after the step.


```python

def gradient(w):
    return 2 * (w - 5)

w = 0

lr = 0.1


grad = gradient(w)

w_new = w - lr * grad

print("Gradient at w=0:", grad)
print("Updated w after one step:", w_new)
```

4. Create a Python script that demonstrates how different learning rates (0.01, 0.1, 0.5) affect the convergence of gradient descent for f(x) = (x-3)**2. Plot the value of x over 20 iterations for each learning rate using matplotlib.<br><br><em><strong>Hint:</strong> Use a loop to update x for each learning rate and plot all three on the same graph.</em>


```python
import numpy as np
import matplotlib.pyplot as plt


def gradient(x):
    return 2 * (x - 3)


learning_rates = [0.01, 0.1, 0.5]


history = {}

iterations = 20

for lr in learning_rates:
    x = 0  # starting point
    values = []

    for _ in range(iterations):
        values.append(x)
        x = x - lr * gradient(x)

    history[lr] = values


plt.figure(figsize=(8, 5))

for lr in learning_rates:
    plt.plot(range(iterations), history[lr], label=f"LR = {lr}")

plt.title("Effect of Learning Rate on Gradient Descent Convergence")
plt.xlabel("Iterations")
plt.ylabel("x value")
plt.legend()
plt.grid(True)

plt.show()
```

5. Write a short Python function that generates and plots a graph of f(x) = x**4 - 3x**3 + 2 for x in range(-2, 4), and visually identify one local minimum and the global minimum on the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**4 - 3*x**3 + 2

x = np.linspace(-2, 4, 400)
y = f(x)

crit_points = np.array([0, 9/4])
crit_values = f(crit_points)


global_min_idx = np.argmin(y)
global_min_x = x[global_min_idx]
global_min_y = y[global_min_idx]


plt.figure(figsize=(8, 5))
plt.plot(x, y, label="f(x) = x^4 - 3x^3 + 2")


plt.scatter(crit_points, crit_values, color='orange', label="Critical Points")


plt.scatter(global_min_x, global_min_y, color='red', label="Global Minimum")

plt.title("Function and Minima Visualization")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)

plt.show()
```
