2. Build a simple LSTM model in Keras to predict the next value in your prepared Flipkart price sequence and train it for 10 epochs.


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
```

```python
prices = np.array([999, 1050, 1020, 1100, 1080, 1150, 1175, 1200, 1190, 1250])
```

```python
window_size = 3

X, y = [], []

for i in range(len(prices) - window_size):
    X.append(prices[i:i+window_size])
    y.append(prices[i+window_size])

X = np.array(X)
y = np.array(y)

X = X.reshape((X.shape[0], X.shape[1], 1))

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
model = Sequential()

model.add(LSTM(50, activation='tanh', input_shape=(3, 1)))
model.add(Dense(1)) 

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X, y, epochs=10, verbose=1)
```

```python
test_input = np.array([1200, 1190, 1250])
test_input = test_input.reshape((1, 3, 1))

predicted_price = model.predict(test_input)
print("Predicted next price:", predicted_price[0][0])
```

4. Write a short code snippet to switch your LSTM layer to a GRU layer in the same model and retrain it. Compare the training time and prediction result with your LSTM model.<br><br><em><strong>Hint:</strong> Replace keras.layers.LSTM with keras.layers.GRU and keep other parameters the same.</em>


```python
import numpy as np
import time
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense  # LSTM → GRU

prices = np.array([999, 1050, 1020, 1100, 1080, 1150, 1175, 1200, 1190, 1250])

window_size = 3
X, y = [], []

for i in range(len(prices) - window_size):
    X.append(prices[i:i+window_size])
    y.append(prices[i+window_size])

X = np.array(X).reshape(-1, 3, 1)
y = np.array(y)

model_gru = Sequential([
    GRU(50, activation='tanh', input_shape=(3, 1)),
    Dense(1)
])

model_gru.compile(optimizer='adam', loss='mse')
```

```python
start = time.time()

model_gru.fit(X, y, epochs=10, verbose=1)

gru_time = time.time() - start
print("GRU Training Time:", gru_time)
```

```python
test_input = np.array([1200, 1190, 1250]).reshape(1, 3, 1)

gru_pred = model_gru.predict(test_input)
print("GRU Prediction:", gru_pred[0][0])
```

| Model | Training Speed | Behavior              | Prediction Difference         |
| ----- | -------------- | --------------------- | ----------------------------- |
| LSTM  | Slower         | Better at long memory | Slightly smoother             |
| GRU   | Faster         | Simpler gates         | Similar or slightly different |


5. Use ChatGPT to generate a Python function that prepares any sequence data (such as daily Spotify song play counts) for LSTM input, and explain in your own words how the function works

```python
import numpy as np

def prepare_lstm_data(sequence, window_size=3):


    X, y = [], []

    for i in range(len(sequence) - window_size):
        X.append(sequence[i:i + window_size])
        y.append(sequence[i + window_size])

    X = np.array(X)
    y = np.array(y)

    X = X.reshape(X.shape[0], X.shape[1], 1)

    return X, y
```

```python
plays = [120, 135, 150, 160, 155, 170, 180]

X, y = prepare_lstm_data(plays, window_size=3)

print("X shape:", X.shape)
print("y shape:", y.shape)
print("First input sample:\n", X[0])
```
