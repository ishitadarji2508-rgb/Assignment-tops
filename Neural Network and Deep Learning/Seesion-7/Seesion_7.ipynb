1. Write a Python function calculate_mse(predictions, targets) that takes two lists of numbers and returns the Mean Squared Error between them using the MSE formula.


```python
def calculate_mse(predictions, targets):
   
    if len(predictions) != len(targets):
        raise ValueError("Predictions and targets must have the same length")
    
    n = len(predictions)
    error_sum = 0

    for i in range(n):
        error = predictions[i] - targets[i]
        error_sum += error ** 2

    mse = error_sum / n
    return mse

predictions = [2.5, 0.0, 2.1, 7.8]
targets = [3.0, -0.1, 2.0, 8.0]

result = calculate_mse(predictions, targets)
print("MSE:", result)
```

2. Given two arrays: actual_ratings = [4, 3, 5, 2, 1] and predicted_ratings = [5, 2, 4, 2, 1], calculate the Mean Absolute Error (MAE) using Python and print the result.

```python
def calculate_mae(actual, predicted):
    if len(actual) != len(predicted):
        raise ValueError("Lists must have the same length")
    
    n = len(actual)
    total_error = 0

    for i in range(n):
        total_error += abs(actual[i] - predicted[i])

    mae = total_error / n
    return mae

actual_ratings = [4, 3, 5, 2, 1]
predicted_ratings = [5, 2, 4, 2, 1]

result = calculate_mae(actual_ratings, predicted_ratings)

print("Mean Absolute Error (MAE):", result)
```

3. Implement a function binary_cross_entropy(y_true, y_pred) in Python that computes the binary cross-entropy loss for a batch of predictions and true labels (both as lists of 0s and 1s), then test it with y_true = [1, 0, 1] and y_pred = [0.9, 0.2, 0.7].

```python
import numpy as np

def binary_cross_entropy(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)

    loss = -np.mean(
        y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)
    )

    return loss

y_true = [1, 0, 1]
y_pred = [0.9, 0.2, 0.7]

result = binary_cross_entropy(y_true, y_pred)

print("Binary Cross-Entropy Loss:", result)
```

4. Simulate a basic Categorical Cross-Entropy calculation: given y_true = [0, 0, 1] and y_pred = [0.2, 0.3, 0.5], write Python code to compute the categorical cross-entropy loss for this single data point.

```python
import numpy as np

def categorical_cross_entropy(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)

    loss = -np.sum(y_true * np.log(y_pred))
    return loss



y_true = [0, 0, 1]
y_pred = [0.2, 0.3, 0.5]

result = categorical_cross_entropy(y_true, y_pred)

print("Categorical Cross-Entropy Loss:", result)
```

Example App: Zomato (Food Recommendation / Rating System)
Scenario

Zomato wants to build a model that predicts how much a user will like a restaurant on a 1–5 rating scale based on past orders, cuisine preference, and reviews.

 Best Loss Function: Mean Squared Error (MSE)
Why MSE is suitable:
The output is a continuous numeric value (1–5 rating), not a class label.
MSE penalizes large errors more heavily, which is useful when predicting ratings because:
Predicting 5 when actual is 1 is a big mistake
Predicting 4 when actual is 5 is a small mistake
It encourages the model to be precise in numerical prediction.
Alternative comparison:
MAE (Mean Absolute Error):
Could also work, but treats all errors equally, so it is less sensitive to large mistakes compared to MSE.
Binary Cross-Entropy:
Not suitable because the output is not binary (0/1).
Categorical Cross-Entropy:
Not suitable because ratings are not categories but ordered numeric values.
📌 Real-world interpretation

Zomato uses such models to:

Predict expected rating for unseen restaurants
Rank restaurants in search results
Personalize recommendations

A well-trained MSE-based model ensures that predicted ratings closely match real user experience, improving recommendation quality and user satisfaction.
