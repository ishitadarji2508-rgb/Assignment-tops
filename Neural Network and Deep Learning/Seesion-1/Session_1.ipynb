4. Write a step-by-step workflow for building a deep learning model for movie recommendation, starting from collecting user data to deploying the trained model in a real app.<br><br><em><strong>Hint:</strong> Use the stages: Data → Model → Training → Evaluation → Deployment.</em>

| Stage                           | Steps                                                                                                                                                                                         |
| ------------------------------- | --------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------- |
| **1. Data Collection**          | Collect user data such as movie ratings, watch history, likes/dislikes, search history, and demographic information. Gather movie information like genre, actors, director, and release year. |
| **2. Data Preprocessing**       | Remove duplicate or missing values. Convert categorical data (genres, actors) into numerical format. Normalize ratings and create user-item interaction matrices.                             |
| **3. Feature Engineering**      | Extract useful features from user behavior and movie metadata. Create embeddings for users and movies to capture preferences and similarities.                                                |
| **4. Model Design**             | Choose a deep learning architecture such as Neural Collaborative Filtering (NCF), Autoencoders, or Deep Neural Networks (DNNs). Define input, hidden, and output layers.                      |
| **5. Training the Model**       | Split data into training and validation sets. Feed user and movie features into the model. Use a loss function (e.g., Mean Squared Error) and an optimizer (e.g., Adam) to learn patterns.    |
| **6. Model Evaluation**         | Test the model on unseen data. Measure performance using metrics such as RMSE, Precision@K, Recall@K, and Accuracy. Compare results with baseline models.                                     |
| **7. Hyperparameter Tuning**    | Adjust learning rate, batch size, number of layers, and neurons to improve recommendation quality.                                                                                            |
| **8. Model Deployment**         | Save the trained model and deploy it using APIs (Flask, FastAPI, or Django). Integrate it into a movie app or website.                                                                        |
| **9. Real-Time Recommendation** | When a user logs in, collect their latest activity and generate personalized movie recommendations instantly.                                                                                 |
| **10. Monitoring & Updates**    | Track model performance, collect new user feedback, and periodically retrain the model to keep recommendations relevant.                                                                      |
