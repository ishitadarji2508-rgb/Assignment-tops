1. Create a Python list representing the daily step count for a week (7 numbers), and explain why a sequence model like an RNN would be more suitable than a standard feedforward neural network for predicting tomorrow's steps.

```python
weekly_steps = [6500, 7200, 8100, 7500, 9000, 8500, 10000]

print("Weekly Steps:", weekly_steps)
```

Why an RNN is More Suitable

A Recurrent Neural Network (RNN) is better for predicting tomorrow's step count because step data is a sequence where the order of days matters. RNNs have memory and can use information from previous days to make predictions.

A standard feedforward neural network treats each input independently and does not naturally remember past values or capture temporal patterns. In contrast, an RNN can learn trends such as increasing activity over the week, weekend behavior, or recurring patterns in daily steps.

Example: If your steps have been increasing steadily for several days, an RNN can recognize this trend and use it to predict tomorrow's steps more accurately.

Implement a simple RNN cell from scratch in Python that takes a sequence of 5 numbers (e.g., song play counts per day from Spotify) and outputs the final hidden state after processing the sequence.<br><br><em><strong>Hint:</strong> Use the formula: h_t = tanh(Wx_t + Uh_{t-1} + b), where h_0 = 0.</em>


```python
import numpy as np

sequence = [100, 120, 150, 130, 170]


W = 0.5   
U = 0.8   
b = 0.1  

h = 0.0

print("Processing sequence:")

for t, x in enumerate(sequence, start=1):
    h = np.tanh(W * x + U * h + b)
    print(f"Step {t}: x={x}, h={h:.4f}")

print("\nFinal Hidden State:", h)
```

3. Given the following sentence: 'I love cricket and IPL matches', convert each word to a unique integer (word-to-index mapping), then simulate how an RNN would process this sequence by updating and printing the hidden state at each step.


```python
import numpy as np

sentence = "I love cricket and IPL matches"

words = sentence.split()

word_to_index = {word: idx + 1 for idx, word in enumerate(words)}

print("Word-to-Index Mapping:")
print(word_to_index)

sequence = [word_to_index[word] for word in words]
print("\nInteger Sequence:")
print(sequence)


W = 0.5  
U = 0.8  
b = 0.1   

h = 0.0

print("\nRNN Hidden State Updates:")

for t, x in enumerate(sequence, start=1):
    h = np.tanh(W * x + U * h + b)
    print(f"Step {t}: Input={x}, Hidden State={h:.4f}")

print("\nFinal Hidden State:", round(h, 4))
```

4. Explain the vanishing gradient problem in RNNs using your own words, and give one real-world example (from text, speech, or time-series apps) where this issue might cause poor predictions.


The vanishing gradient problem occurs when an RNN is trained on long sequences. During backpropagation, the gradients (which tell the network how to update its weights) are multiplied many times as they travel backward through time. These gradients can become extremely small, eventually approaching zero.

When this happens, the network learns very little from earlier time steps because the weight updates become negligible. As a result, the RNN struggles to remember information from the distant past and focuses mostly on recent inputs.

Real-World Example

Consider a movie review sentiment analysis task:

"Although the movie started brilliantly, had great acting, and an interesting story, the ending was disappointing."

To correctly classify the sentiment, the model should remember important words from the beginning of the sentence such as "brilliant" and "great acting". With the vanishing gradient problem, a basic RNN may forget these earlier words by the time it reaches the end of the sentence and may base its prediction mainly on "disappointing", leading to an inaccurate sentiment prediction.

Why It Matters
Poor memory of long-term dependencies.
Reduced accuracy on long sequences.
Difficulty learning patterns that depend on information far back in the sequence.
Common Solution

Modern architectures such as the entity reference is not appropriate here; use plain text LSTM (Long Short-Term Memory) and GRU (Gated Recurrent Unit) networks use gating mechanisms to preserve important information over longer periods, helping overcome the vanishing gradient problem.

5. Use ChatGPT to generate a short Python code snippet that demonstrates Backpropagation Through Time (BPTT) for a 3-step RNN sequence. Paste the code and briefly describe what each part does

```python
import numpy as np

x = [1.0, 2.0, 3.0]


W = 0.5     
U = 0.8     
b = 0.1     


h = [0.0]

for t in range(3):
    h_t = np.tanh(W * x[t] + U * h[-1] + b)
    h.append(h_t)

target = 0.5
loss = 0.5 * (h[3] - target) ** 2

print("Loss:", loss)

dW = 0.0
dU = 0.0

dh = h[3] - target

for t in reversed(range(1, 4)):
    dtanh = dh * (1 - h[t] ** 2)

    dW += dtanh * x[t - 1]
    dU += dtanh * h[t - 1]


    dh = dtanh * U

print("Gradient dW:", dW)
print("Gradient dU:", dU)
```
