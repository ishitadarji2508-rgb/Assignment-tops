1. Build a simple movie recommendation neural network using Keras Functional API that takes two inputs: user age and favorite genre (as integers), and outputs a predicted movie rating between 1 and 5.


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

age_input = keras.Input(shape=(1,), name="age")
genre_input = keras.Input(shape=(1,), name="genre")

age_branch = layers.Dense(16, activation="relu")(age_input)
genre_branch = layers.Dense(16, activation="relu")(genre_input)

combined = layers.concatenate([age_branch, genre_branch])

x = layers.Dense(32, activation="relu")(combined)
x = layers.Dense(16, activation="relu")(x)

output = layers.Dense(1, activation="linear")(x)

model = keras.Model(inputs=[age_input, genre_input], outputs=output)

model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

model.summary()
```

2. Create a multi-output neural network using Keras Functional API that predicts both the delivery time (in minutes) and delivery rating (1-5 stars) for a Zomato-style food order, given order amount and distance as inputs.<br><br><em><strong>Hint:</strong> Use two Dense output layers, one for each prediction.</em>


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

order_amount = keras.Input(shape=(1,), name="order_amount")
distance = keras.Input(shape=(1,), name="distance")

x1 = layers.Dense(16, activation="relu")(order_amount)
x2 = layers.Dense(16, activation="relu")(distance)

combined = layers.concatenate([x1, x2])

x = layers.Dense(32, activation="relu")(combined)
x = layers.Dense(16, activation="relu")(x)

delivery_time = layers.Dense(1, name="delivery_time")(x)
rating = layers.Dense(1, name="delivery_rating")(x)

model = keras.Model(
    inputs=[order_amount, distance],
    outputs=[delivery_time, rating]
)

model.compile(
    optimizer="adam",
    loss={
        "delivery_time": "mse",
        "delivery_rating": "mse"
    },
    metrics={
        "delivery_time": "mae",
        "delivery_rating": "mae"
    }
)

model.summary()
```

3. Modify an existing Keras Sequential model for music playlist recommendations to use the Functional API instead, allowing for future flexibility such as adding more inputs or outputs.<br><br><em><strong>Hint:</strong> Focus on rewriting the model definition, not the data pipeline.</em>


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

inputs = keras.Input(shape=(2,), name="user_song_features")

x = layers.Dense(16, activation="relu")(inputs)
x = layers.Dense(8, activation="relu")(x)

output = layers.Dense(1, activation="sigmoid", name="recommendation")(x)

model = keras.Model(inputs=inputs, outputs=output)

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

4. mplement a skip connection in a Keras Functional API model for an IPL cricket score predictor, where the input layer connects both to an intermediate Dense layer and directly to the output layer.<br><br><em><strong>Hint:</strong> Use the 'Add' layer to combine outputs from different layers.</em>

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

inputs = keras.Input(shape=(4,), name="match_features")

x = layers.Dense(32, activation="relu")(inputs)

x = layers.Dense(32, activation="relu")(x)

skip = layers.Dense(32, activation="linear")(inputs)

x = layers.Add()([x, skip])
x = layers.Activation("relu")(x)

output = layers.Dense(1, name="predicted_score")(x)

model = keras.Model(inputs=inputs, outputs=output)

model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

model.summary()
```
