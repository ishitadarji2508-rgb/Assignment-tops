4. Compile your model using the 'adam' optimizer, 'categorical_crossentropy' loss, and track 'accuracy' as a metric. Then, train the model for 5 epochs using random data shaped like CIFAR-10 images and labels.<br><br><em><strong>Constraint:</strong> Use numpy to generate random input data and one-hot encoded labels for testing if you don't want to download the full CIFAR-10 dataset.</em>


```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    BatchNormalization,
    MaxPooling2D,
    Flatten,
    Dropout,
    Dense
)
from tensorflow.keras.utils import to_categorical

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

num_samples = 1000

X_train = np.random.rand(num_samples, 32, 32, 3).astype(np.float32)

y_train = np.random.randint(0, 10, size=(num_samples,))
y_train = to_categorical(y_train, num_classes=10)

history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.2
)

loss, accuracy = model.evaluate(X_train, y_train)

print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Conv2D,
    BatchNormalization,
    MaxPooling2D,
    Flatten,
    Dropout,
    Dense
)
from tensorflow.keras.utils import to_categorical

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    Conv2D(64, (3, 3), activation='relu'),
    BatchNormalization(),
    MaxPooling2D((2, 2)),

    Flatten(),
    Dropout(0.3),

    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

num_samples = 1000

X_train = np.random.rand(num_samples, 32, 32, 3).astype(np.float32)

y_train = np.random.randint(0, 10, size=num_samples)
y_train = to_categorical(y_train, num_classes=10)

history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_split=0.2
)

loss, accuracy = model.evaluate(X_train, y_train)

print(f"Loss: {loss:.4f}")
print(f"Accuracy: {accuracy:.4f}")
```
