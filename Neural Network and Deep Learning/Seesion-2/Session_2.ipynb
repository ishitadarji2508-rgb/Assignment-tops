1. Create a Python script using NumPy to implement a perceptron that learns to classify whether a Flipkart product review is 'positive' (1) or 'negative' (0) based on two features: review length and number of exclamation marks. Use a small manually created dataset for training.

```python
import numpy as np
X = np.array([
    [100, 5],   
    [80, 4],   
    [20, 0],    
    [30, 1]     
])


y = np.array([1, 1, 0, 0])

weights = np.zeros(X.shape[1])
bias = 0
learning_rate = 0.01


for epoch in range(20):
    for i in range(len(X)):
        linear_output = np.dot(X[i], weights) + bias
        prediction = 1 if linear_output >= 0 else 0

        error = y[i] - prediction

        weights += learning_rate * error * X[i]
        bias += learning_rate * error

print("Final Weights:", weights)
print("Final Bias:", bias)


for sample in X:
    output = 1 if np.dot(sample, weights) + bias >= 0 else 0
    print(f"Input: {sample}, Prediction: {output}")
```

Test your function with sample weights, inputs, and bias values, and print the output for each case.<br><br><em><strong>Hint:</strong> Use NumPy arrays for vectorized operations.</em>


```python
import numpy as np

def perceptron(inputs, weights, bias):
    weighted_sum = np.dot(inputs, weights) + bias

    
    output = 1 if weighted_sum >= 0 else 0

    return weighted_sum, output

tests = [
    (np.array([2, 3]), np.array([0.5, 0.2]), -1),
    (np.array([1, 1]), np.array([0.4, 0.4]), -0.5),
    (np.array([4, 2]), np.array([0.3, 0.6]), -2)
]

for x, w, b in tests:
    ws, out = perceptron(x, w, b)
    print("Inputs:", x)
    print("Weighted Sum:", ws)
    print("Output:", out)
```

4. Implement a simple Multilayer Perceptron (MLP) in Python using only NumPy that can solve the XOR problem. Your network should have an input layer, one hidden layer, and an output layer, with manual forward propagation (no external libraries).<br><br><em><strong>Hint:</strong> Use sigmoid activation in the hidden and output layers.</em>


```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# XOR Inputs
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

W1 = np.array([
    [20, 20],
    [20, 20]
])

b1 = np.array([-10, -30])

W2 = np.array([
    [20],
    [-20]
])

b2 = np.array([-10])

for x in X:

    hidden_input = np.dot(x, W1) + b1
    hidden_output = sigmoid(hidden_input)

    final_input = np.dot(hidden_output, W2) + b2
    final_output = sigmoid(final_input)

    prediction = 1 if final_output >= 0.5 else 0

    print(
        f"Input={x}, "
        f"Output={final_output[0]:.4f}, "
        f"Prediction={prediction}"
    )
```

5. Modify your MLP code to classify whether a Zomato restaurant is 'trending' (1) or 'not trending' (0) based on two input features: average rating and number of recent reviews. Create a tiny dataset and run forward propagation to show the output for each sample.

```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

X = np.array([
    [4.8, 300],   
    [4.5, 250],   
    [3.2, 50],    
    [3.8, 70]     
])

W1 = np.array([
    [0.5, 0.3],
    [0.01, 0.02]
])

b1 = np.array([0.1, 0.1])

W2 = np.array([
    [1.0],
    [1.2]
])

b2 = np.array([-1])

for sample in X:

    hidden = sigmoid(np.dot(sample, W1) + b1)

    output = sigmoid(np.dot(hidden, W2) + b2)

    prediction = 1 if output >= 0.5 else 0

    print(
        f"Restaurant={sample}, "
        f"Score={output[0]:.4f}, "
        f"Prediction={prediction}"
    )
```
