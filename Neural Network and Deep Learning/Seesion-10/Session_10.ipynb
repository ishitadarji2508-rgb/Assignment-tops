1. Create a simple Keras Sequential model for classifying handwritten digits (use the MNIST dataset), and train it without any regularization. Save the validation accuracy after 5 epochs.


```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax")
])

model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

history = model.fit(
    x_train,
    y_train,
    epochs=5,
    validation_data=(x_test, y_test),
    verbose=1
)

val_accuracy_after_5_epochs = history.history["val_accuracy"][-1]
print(val_accuracy_after_5_epochs)
```

2. Modify your Keras model to add L2 regularization (kernel_regularizer=regularizers.l2(0.01)) to all Dense layers, retrain for 5 epochs, and compare the validation accuracy to your previous result.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, regularizers

mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation="relu", kernel_regularizer=regularizers.l2(0.01)),
    layers.Dense(10, activation="softmax", kernel_regularizer=regularizers.l2(0.01))
])

model.compile(optimizer="adam",
              loss="sparse_categorical_crossentropy",
              metrics=["accuracy"])

history = model.fit(
    x_train,
    y_train,
    epochs=5,
    validation_data=(x_test, y_test),
    verbose=1
)

val_accuracy_after_5_epochs = history.history["val_accuracy"][-1]
print(val_accuracy_after_5_epochs)
```

3. Add Dropout layers (with rate=0.5) after each Dense layer in your Keras model, retrain for 5 epochs, and observe the change in validation accuracy.<br><br><em><strong>Hint:</strong> Use keras.layers.Dropout and place it after activations.</em>

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation="relu"),
    layers.Dropout(0.5),
    layers.Dense(10, activation="softmax"),
    layers.Dropout(0.5)
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history = model.fit(
    x_train,
    y_train,
    epochs=5,
    validation_data=(x_test, y_test),
    verbose=1
)

val_accuracy_after_5_epochs = history.history["val_accuracy"][-1]
print(val_accuracy_after_5_epochs)
```

4. Integrate BatchNormalization layers into your Keras model (before activation functions in Dense layers), retrain, and note how the training and validation accuracy curves change.<br><br><em><strong>Hint:</strong> Use keras.layers.BatchNormalization before the activation in each Dense block.</em>

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),

    layers.Dense(128, use_bias=False),
    layers.BatchNormalization(),
    layers.Activation("relu"),

    layers.Dense(10, use_bias=False),
    layers.BatchNormalization(),
    layers.Activation("softmax")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

history = model.fit(
    x_train,
    y_train,
    epochs=5,
    validation_data=(x_test, y_test),
    verbose=1
)

print("Final training accuracy:", history.history["accuracy"][-1])
print("Final validation accuracy:", history.history["val_accuracy"][-1])
```

5. Implement EarlyStopping in your model training using keras.callbacks.EarlyStopping with patience=2, and show how it affects the number of training epochs and final validation accuracy.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

mnist = keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype("float32") / 255.0

model = keras.Sequential([
    layers.Flatten(input_shape=(28, 28)),
    layers.Dense(128, activation="relu"),
    layers.Dense(10, activation="softmax")
])

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

early_stop = keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=2,
    restore_best_weights=True
)

history = model.fit(
    x_train,
    y_train,
    epochs=20,
    validation_data=(x_test, y_test),
    callbacks=[early_stop],
    verbose=1
)

print("Total epochs run:", len(history.history["loss"]))
print("Final validation accuracy:", history.history["val_accuracy"][-1])
```
